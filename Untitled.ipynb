```python
def trydate(x):
    x = x[6:-7]
    try:
        return datetime.fromtimestamp(int(x) / 1000).date()
    except:
        print(x)
        raise
br_data["date"] = br_data["EventDate"].str[6:-7].apply(trydate)
```

```python
import time
import requests

br_url = "https://www.bikereg.com/api/search"

def load_br_data(startpage=0):
    br_page_url = br_url + "?startpage={}".format(startpage)
    rj = requests.get(br_page_url).json()["MatchingEvents"]
    return pd.DataFrame(rj)

responses = []
latest_df = ['x']
i = 0
while len(latest_df) > 0:
    print("loading page {}".format(i))
    latest_df = load_br_data(startpage=i)
    print("{} results".format(len(latest_df)))
    responses.append(latest_df)
    print("====")
    time.sleep(1)
    i = i + 1
```

```python
all_events = pd.concat(responses, ignore_index=True)
```

```python
all_events["date"] = all_events["EventDate"].str[6:-7].apply(trydate)
```
